# Channel selection procedure plots

```python
import numpy as np
import matplotlib.pyplot as plt
from dim_reduction import select_best_chanels
```

```python
def plot_chanels_knn_acc(subjets_acc):
    number_of_chanels = np.arange(1, len(subjets_acc)+1)
    mean_subject_acc = [sum(humans_acc)/len(humans_acc) for humans_acc in subjets_acc]
    
    
    fig, ax = plt.subplots()
    ax.plot(number_of_chanels, mean_subject_acc, label='Mean Accuracy')
    for ch_num in range(len(subjets_acc[0])):
        ax.plot(number_of_chanels, [person_acc[ch_num] for person_acc in subjets_acc], label='volunteer _'+str(ch_num+1))
        
    ax.set(xlabel='Number of channels', ylabel='Accuracy',
           title='Channel selection')
    ax.legend(loc='upper right')
    
subset_acc, chanels_rank = select_best_chanels()
plot_chanels_knn_acc(subset_acc)  
```
